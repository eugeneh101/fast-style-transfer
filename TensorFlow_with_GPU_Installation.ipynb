### Mixed the instructions from the following tutorials to get TensorFlow 1.7.0 with CUDA 9.0 correctly installed  
Youtube instructions are the most update to date compared to articles:  
https://www.youtube.com/watch?v=6iyweMKcX3w
https://www.youtube.com/watch?v=qESp8mSVmTM
https://blog.paperspace.com/running-tensorflow-on-windows/  
https://medium.com/@raza.shahzad/setting-up-tensorflow-gpu-keras-in-conda-on-windows-10-75d4fd498198  
https://medium.com/@viveksingh.heritage/how-to-install-tensorflow-gpu-version-with-jupyter-windows-10-in-8-easy-steps-8797547028a  
Windows only uses CUDA 9.0 for TensorFlow 1.7 (needs associated upgrade packs and cuDNN) or can use CUDA 8.0; need to make sure path variables are correct. Basically Windows installation was too time consuming, so I got GPU acceleration installed on Ubuntu instead.    

````python
### Copy and paste this to shell connected to AWS EC2 instance
wget https://s3.amazonaws.com/style-transfer-installation/Installing_TensorFlow_with_CUDA.sh
time bash Installing_TensorFlow_with_CUDA.sh
# For debugging: tail --lines=+40 Installing_TensorFlow_with_CUDA.sh | head | bash 2> err_log

# After entire script runs, it is best to reboot the instance: "sudo reboot"
# Remember to turn on the conda env: "source activate tensorflow"
````
